```python
from scipy.stats import chi2_contingency
```

```python
data = [[11, 3, 1, 5], [33, 56, 15, 6]]
stat, p, dof, expected = chi2_contingency(data)
  
# interpret p-value
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```

    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
